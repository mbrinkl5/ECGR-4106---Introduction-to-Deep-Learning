## Overview

- **Problem 1:** Vision Transformer from Scratch trained on CIFAR-100, compared against ResNet18 baseline.
- **Problem 2:**  Fine-tune two pretrained Swin Transformer models on the CIFAR-100 dataset, compare against a Swin Transformer from scratch.

## Setup

All experiments were implemented in PyTorch and trained on Google Colab using an  A100 GPU with a fixed random seed of 42.

**Problem 1 (Vision Transformer):**
- Optimizer: Adam
- Learning rate: 0.001
- Batch size: 64
- Epochs: 10
- Dataset: CIFAR-100 (32×32 RGB images)

**Problem 2 (Swin Transformer):**
- Optimizer: Adam
- Pretrained Swin-Tiny and Swin-Small: learning rate 2e-5, batch size 32, 5 epochs (classification head fine-tuned only)
- Scratch Swin Transformer: learning rate 0.001, batch size 32, 5 epochs
- Dataset: CIFAR-100 resized to 224×224 using ImageNet normalization
